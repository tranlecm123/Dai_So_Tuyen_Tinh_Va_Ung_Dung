```python

#TranVanLe-2374802010270
#1. Dẫn nhập – Một số hàm về xử lý vector với Python 
# Hàm scale: Nhân mỗi phần tử trong vector với một số thực
def scale(a, v):
    return [a * vi for vi in v]

# Ví dụ:
v = [3, 5, 7]
print("scale(10, v):", scale(10, v))

```

```python
# Hàm cộng hai vector
def sumvector(v, w):
    return [vi + wi for (vi, wi) in zip(v, w)]

# Ví dụ:
v = [3, 5, 7]
w = [2, 4, 6]
print("sumvector(v, w):", sumvector(v, w))

```

```python
# Hàm tích vô hướng giữa hai vector
def dotvector(v, w):
    return sum([vi * wi for (vi, wi) in zip(v, w)])

# Ví dụ:
print("dotvector(v, w):", dotvector(v, w))
```

```python
# Dùng lại hàm dotvector để tính độ dài bình phương
def lenvector(v):
    return dotvector(v, v)

# Ví dụ:
print("lenvector(w):", lenvector(w))
```

```python
#2. Bài toán ứng dụng 1 – Phân loại tuyến tính  

import numpy as np

# Mảng điểm số từ tích vô hướng của w và x
scores = np.array([-1, 1, 2, -3, 5, -4])
print("Scores:", scores)

# Kiểm tra phần tử nào >= 0
print("scores >= 0:", scores >= 0)

# Kiểm tra phần tử nào < 0
print("scores < 0:", scores < 0)

# Phân loại: >=0 là "so duong", <0 là "so am"
labels = np.select([scores >= 0, scores < 0], ['so duong', 'so am'])
print("Phân loại nhị phân:", labels)
```

```python
#3. Thực hành xử lý ma trận
# Phần 1: Khai báo và thao tác cơ bản với ma trận

import numpy as np
from scipy import linalg

# 1. Khai báo ma trận thủ công với np.mat
D = np.mat([[3, 4], [5, 6]])
print("Ma trận D:")
print(D)
print()

# 2. Tạo ma trận ngẫu nhiên kích thước 5x7
C = np.mat(np.random.random((5, 7)))
print("Ma trận ngẫu nhiên C:")
print(C)
print()

# 3. Tạo ma trận vuông ngẫu nhiên 2x2
A = np.mat(np.random.random((2, 2)))
print("Ma trận vuông A:")
print(A)
print()

# 4. Chuyển đổi array sang ma trận
b = np.array([(1+5j, 2j, 3j), (4, 5, 6)])
B = np.asmatrix(b)
print("Mảng b:")
print(b)
print()
print("Ma trận B (từ b):")
print(B)
print()

# 5. Các phép toán cơ bản

# Chuyển vị ma trận A
print("Chuyển vị của A (A.T):")
print(A.T)
print()

# Ma trận nghịch đảo bằng numpy
print("Ma trận nghịch đảo A (sử dụng .I):")
print(A.I)
print()

# Ma trận nghịch đảo bằng scipy.linalg
print("Ma trận nghịch đảo A (sử dụng linalg.inv):")
print(linalg.inv(A))
print()

# 6. Ma trận tam giác và vector đường chéo
M = np.array([[-1, 3, 2], [0, -2, 1], [1, 5, -2]])
print("Ma trận M:")
print(M)
print()

print("Ma trận tam giác dưới của M (np.tril):")
print(np.tril(M))
print()

print("Ma trận tam giác trên của M (np.triu):")
print(np.triu(M))
print()

v_diag = np.diag(M)  # vector đường chéo của M
print("Vector đường chéo của M:")
print(v_diag)
print()

M_diag = np.diag(v_diag)  # ma trận đường chéo từ vector v_diag
print("Ma trận đường chéo từ vector đường chéo:")
print(M_diag)
print()

# 7. Ma trận đơn vị kích thước 2x2
I2 = np.identity(2)
print("Ma trận đơn vị I2:")
print(I2)
print()
```

```python
# Phần 2: Giải hệ phương trình với SymPy và so sánh ma trận

import sympy as sp
import numpy as np

# 1. Khai báo biến
x, y, z = sp.symbols('x y z')

# 2. Đặt hệ phương trình từ đề bài (2 ma trận bằng nhau)
equations = [
    x**2 + 2 - 6,     # A[0][1] = B[0][1]
    y - 1 - 6,        # A[1][0] = B[1][0]
    x + z - 1,        # A[1][2] = B[1][2]
    2*x**2 - 8,       # A[2][0] = B[2][0]
    z**2 + 4 - 5,     # A[2][1] = B[2][1]
    y - z - 8         # A[2][2] = B[2][2]
]

# 3. Giải hệ phương trình
solution = sp.solve(equations, (x, y, z))
print("Nghiệm tìm được:")
print(solution)
print()

# 4. Kiểm tra lại nghiệm (lấy nghiệm đầu tiên nếu có nhiều nghiệm)
if solution:
    x_val, y_val, z_val = solution[0][x], solution[0][y], solution[0][z]

    print(f"Kiểm tra nghiệm với x={x_val}, y={y_val}, z={z_val}:")

    print("x^2 + 2 =", x_val**2 + 2)
    print("y - 1 =", y_val - 1)
    print("x + z =", x_val + z_val)
    print("2*x^2 =", 2 * x_val**2)
    print("z^2 + 4 =", z_val**2 + 4)
    print("y - z =", y_val - z_val)
    print()

# 5. So sánh 2 list trong Python thuần
list1 = [1, 2, 3]
list2 = [1, 2, 3]
print("So sánh 2 list bằng all():", all([list1[i] == list2[i] for i in range(len(list1))]))

list3 = [1, 2, 4]
print("So sánh 2 list khác nhau bằng all():", all([list1[i] == list3[i] for i in range(len(list1))]))
print()

# 6. So sánh 2 mảng numpy
print("So sánh 2 mảng numpy (array_equal):")
print(np.array_equal(np.array([1, 2]), np.array([1, 2])))  # True
print(np.array_equal(np.array([1, 2]), np.array([1, 3])))  # False
```

```python
#3.2. Các phép biến đổi sơ cấp trên ma trận 
import numpy as np

# Phần 1: Xây dựng ma trận A 6x6 và ma trận đơn vị I6, kiểm tra kích thước và đường chéo
print("=== Phần 1: Xây dựng ma trận A, I6, kiểm tra kích thước và đường chéo ===")
# Xây dựng ma trận A
A = np.reshape(np.arange(36.0), (6,6))
print("Ma trận A:")
print(A)

# Xây dựng ma trận đơn vị I6
I6 = np.identity(6)
print("\nMa trận đơn vị I6:")
print(I6)

# Kích thước ma trận A
print("\nKích thước ma trận A (số lượng phần tử):", A.size)

# Đường chéo chính của ma trận A
print("\nĐường chéo chính của ma trận A:")
print(np.matrix.diagonal(A))

# Thành lập ma trận A mới bằng A + I6
A = A + I6
print("\nMa trận A mới sau khi cộng với I6:")
print(A)

# Phần 2: Tính tích ma trận A với vector vecB
print("\n=== Phần 2: Tính tích ma trận A với vector vecB ===")
# Vector B
vecB = np.array([1., 2., 3., 4., 5., 6.])

# Tính tích A.dot(vecB)
C = A.dot(vecB)
print("Kết quả tích A.dot(vecB):")
print(C)

# Phần 3: Khai báo và in ma trận D 2x6
print("\n=== Phần 3: Khai báo và in ma trận D ===")
# Khai báo ma trận D
D = np.array([[1., 2., 3., 4., 5., 6.], [1., 0., 1., 0., 1., 0.]])
print("Ma trận D:")
print(D)

# Phần 4: Tính tích ma trận A x D
print("\n=== Phần 4: Tính tích ma trận A x D ===")
# Tính tích A.dot(D)
E = A.dot(D.T)  # Chuyển vị D để phù hợp kích thước
print("Kết quả tích A.dot(D):")
print(E)

# Phần 5: Xây dựng ma trận F 6x2 và tính tích A x F
print("\n=== Phần 5: Xây dựng ma trận F và tính tích A x F ===")
# Ma trận F
F = np.array([[1., 1.], [2., 0.], [3., 1.], [4., 0.], [5., 1.], [6., 0.]])

# In ma trận F
print("Ma trận F:")
print(F)

# Tính tích A.dot(F)
G = A.dot(F)
print("\nKết quả tích A.dot(F):")
print(G)

# Phần 6: Tính ma trận nghịch đảo của A
print("\n=== Phần 6: Tính ma trận nghịch đảo của A ===")
# Tính ma trận nghịch đảo
A_inv = np.linalg.inv(A)
print("Ma trận nghịch đảo của A:")
print(A_inv)

# Phần 7: Câu hỏi về lệnh np.linalg.inv(np.linalg.inv(A))
print("\n=== Phần 7: Kiểm tra np.linalg.inv(np.linalg.inv(A)) ===")
# Tính nghịch đảo của nghịch đảo
A_inv_inv = np.linalg.inv(np.linalg.inv(A))
print("Kết quả np.linalg.inv(np.linalg.inv(A)):")
print(A_inv_inv)

# Kiểm tra xem có bằng A ban đầu không
print("\nSo sánh với ma trận A ban đầu:")
print(np.allclose(A, A_inv_inv))
```

```python
#BÀI TẬP CHƯƠNG 2
#Câu 1: Hãy sử dụng numpy để giải các phương trình sau: 
import numpy as np
import sympy as sp

# Bài 1: Tìm điểm giao nhau của hai đường thẳng x - y = -2 và 2x + 3y = 6 trong R^2
A1 = np.array([[1, -1], [2, 3]])
b1 = np.array([-2, 6])
solution1 = np.linalg.solve(A1, b1)
print("Bài 1: Điểm giao nhau (x, y) =", solution1)

# Bài 2: Tìm điểm giao nhau của các mặt phẳng x - y = 2, 2x - y - z = 3, và x + y + z = 6 trong R^3
A2 = np.array([[1, -1, 0], [2, -1, -1], [1, 1, 1]])
b2 = np.array([2, 3, 6])
solution2 = np.linalg.solve(A2, b2)
print("Bài 2: Điểm giao nhau (x, y, z) =", solution2)

# Bài 3: Tìm đa thức bậc hai p(x) = ax^2 + bx + c sao cho p(1) = 4, p(2) = 3, p(3) = 4
A3 = np.array([[1, 1, 1], [4, 2, 1], [9, 3, 1]])
b3 = np.array([4, 3, 4])
solution3 = np.linalg.solve(A3, b3)
print("Bài 3: Hệ số (a, b, c) =", solution3)
print("Bài 3: Đa thức p(x) =", f"{solution3[0]}x^2 + {solution3[1]}x + {solution3[2]}")

# Bài 4: Tính giá trị tích phân ∫ từ 0 đến 1 của hàm x(x-3)/((x-1)^2(x+2)) dx
x = sp.Symbol('x')  # Khai báo biến tượng trưng x
integrand = (x * (x - 3)) / ((x - 1)**2 * (x + 2))  # Hàm cần tích phân
integral = sp.integrate(integrand, (x, 0, 1))  # Tính tích phân định nghĩa từ 0 đến 1
print("Bài 4: Kết quả tích phân =", integral)
```

```python
# Câu 2: Hãy viết các câu lệnh của sympy để giải các phương trình ở Câu 1
import sympy as sp

# Khai báo các biến tượng trưng
x, y, z = sp.symbols('x y z')

#ìm điểm giao nhau của hai đường thẳng x - y = -2 và 2x + 3y = 6 trong R^2
eq1_1 = sp.Eq(x - y, -2)  # Phương trình 1
eq1_2 = sp.Eq(2 * x + 3 * y, 6)  # Phương trình 2
solution1 = sp.solve((eq1_1, eq1_2), (x, y))  # Giải hệ phương trình
print("Bài 1: Điểm giao nhau (x, y) =", solution1)

#Tìm điểm giao nhau của các mặt phẳng x - y = 2, 2x - y - z = 3, và x + y + z = 6 trong R^3
eq2_1 = sp.Eq(x - y, 2)  # Phương trình 1
eq2_2 = sp.Eq(2 * x - y - z, 3)  # Phương trình 2
eq2_3 = sp.Eq(x + y + z, 6)  # Phương trình 3
solution2 = sp.solve((eq2_1, eq2_2, eq2_3), (x, y, z))  # Giải hệ phương trình
print("Câu 2: Điểm giao nhau (x, y, z) =", solution2)
```

```python
import numpy as np

def fibonacci_matrix_power(k):
    if k < 0:
        raise ValueError("k phải là số nguyên không âm")

    if k == 0:
        F_k_matrix = np.array([[1, 0], [0, 1]])
    else:
        Q = np.array([[1, 1], [1, 0]])
        F_k_matrix = np.linalg.matrix_power(Q, k)

    return F_k_matrix

def fibonacci_number(k):
    """
    Tính số Fibonacci thứ k (F(0)=0, F(1)=1) sử dụng ma trận lũy thừa.
    """
    if k == 0:
        return 0
    F_k_matrix = fibonacci_matrix_power(k-1)
    # F(k) = phần tử (0,0) của Q^(k-1)
    return F_k_matrix[0,0]

# --- Ví dụ chạy thử ---
if __name__ == "__main__":
    for k in [5, 1, 0, 2]:
        print(f"--- Ví dụ với k = {k} ---")
        mat = fibonacci_matrix_power(k)
        print(f"Lũy thừa ma trận Q^{k}:\n{mat}")
        if k > 0:
            print(f"Số Fibonacci thứ {k} là: {fibonacci_number(k)}")
        else:
            print("F(0) = 0")
```
