```python
# Tran Van Le - 2374802010270
#1.2.Tích vô hướng, trực giao và ứng dụn
import numpy as np
import math
from scipy import linalg

a = np.array([1,2,3])
b = np.array([4,5,6])

print(np.inner(a,b))
print(np.dot(a,b))

a = np.array([1,2,-3])
print(np.linalg.norm(a))
print(np.linalg.norm(a, 2))
print(np.linalg.norm(a,1))

def chuan(v, k):
    return sum(abs(i)**k for i in v)**(1/k)

print(chuan(a, 1))
print(np.linalg.norm(a,1))
print(chuan(a, 2))
print(np.linalg.norm(a,2))
print(chuan(a, 3))
print(np.linalg.norm(a,3))
print(chuan(a, 1000))
print(np.linalg.norm(a,1000))

print(linalg.norm(a))
print(linalg.norm(a, 100))
print(linalg.norm(a, 1000))

u = np.array([1,2,-3])
v = np.array([4,5,6])
print(np.linalg.norm(u - v))

tu = np.dot(u, v)
mau = np.linalg.norm(u) * np.linalg.norm(v)
goc = math.acos(tu / mau)
print(goc)

if np.dot(u, v) == 0:
    print("u và v trực giao")
else:
    print("u và v không trực giao")

proj = (np.dot(u, v) / np.dot(v, v)) * v
print(proj)
```

```python
#3. Tri riêng, vector riêng của ma trận và ứng dụng 
#3.1.Bài toán dẫn nhập 
import numpy as np
import sympy as sym

# Dữ liệu đề bài
F = 0.136
J = 0.998

# Ma trận chuyển trạng thái A
A = np.array([[J, F],
              [1 - J, 1 - F]])

# Trạng thái ban đầu: 95% có việc, 5% thất nghiệp
v0 = np.array([0.95, 0.05])

# Tính A^100 và trạng thái sau 100 tuần
A100 = np.linalg.matrix_power(A, 100)
vk = A100.dot(v0)

print("Ma trận A^100:")
print(A100)
print("Trạng thái sau 100 tuần (có việc, thất nghiệp):")
print(vk)

# ---------------------------
# Giải trạng thái cân bằng bằng numpy
print("\nTrạng thái cân bằng (giải hệ bằng numpy):")
A_he = np.array([[-0.002, 0.136],
                 [1, 1]])
b_he = np.array([0, 1])
nghiem_np = np.linalg.solve(A_he, b_he)
print(nghiem_np)

# ---------------------------
# Giải hệ bằng sympy
print("\nTrạng thái cân bằng (giải hệ bằng sympy):")
x, y = sym.symbols('x y')
xy = sym.Matrix([x, y])
A_sym = sym.Matrix([[1, -68],
                    [1, 1]])
v_sym = sym.Matrix([0, 1])
nghiem_sym = sym.solve([A_sym * xy - v_sym])
print(sym.pretty(nghiem_sym))
```

```python
#3.2.Trị riêng và vector riêng của ma trận 
import sympy as sym

# Khai báo biến
x, y, Lambda = sym.symbols('x y Lambda')

# =======================
# PHẦN 1 — Kiểm chứng ví dụ 2x2: A = [[1, 6], [5, 2]], u = [6, -5], λ = -4
A1 = sym.Matrix([[1, 6], [5, 2]])
u = sym.Matrix([6, -5])
lambda_val = -4

# Kiểm chứng: A1 * u = -4 * u
print("Kiểm chứng trị riêng:")
print("A1 * u =", A1 * u)
print("λ * u  =", lambda_val * u)

# =======================
# PHẦN 2 — Tính trị riêng của ma trận A = [[2, 3], [3, -6]]
A2 = sym.Matrix([[2, 3], [3, -6]])
I = sym.eye(2)

# Tính trị riêng: giải phương trình det(λI - A) = 0
pt = sym.Eq(sym.det(Lambda * I - A2), 0)
nghiem = sym.solve(pt)

print("\nTrị riêng (giá trị nghiệm lambda):")
print([sym.N(val, 4) for val in nghiem])
print(sym.pretty(nghiem))

# =======================
# PHẦN 3 — Tính vector riêng tương ứng với λ = 3
lambda_3 = 3
A3 = A2 - lambda_3 * I  # (A - λI)
x, y = sym.symbols('x y')
vec = sym.Matrix([x, y])

# Giải hệ (A - λI) * v = 0
he = A3 * vec
nghiem_vec = sym.solve(he, [x, y])

print("\nVector riêng tương ứng với λ = 3:")
print(nghiem_vec)
```

```python
#3.3.Ứng dụng của trị riêng và vector riêng 
import numpy as np
from numpy import linalg as LA

# Ma trận A
A = np.array([[2, 3],
              [3, -6]])

# Ma trận D - ma trận đường chéo gồm các trị riêng (eigenvalues)
D = np.array([[-7, 0],
              [0, 3]])

# Ma trận P - ma trận cột các vector riêng tương ứng
P = np.array([[-1.0/3, 3],
              [1,      1]])

# Tính ma trận nghịch đảo của P
P_inv = LA.inv(P)
print("P^-1 =")
print(P_inv)

# Kiểm tra A.dot(P) có bằng P.dot(D) không
print("A.dot(P) =")
print(A.dot(P))

print("P.dot(D) =")
print(P.dot(D))

# Tính lũy thừa A^1000 = P·D^1000·P^{-1}
A_1000 = P.dot(np.linalg.matrix_power(D, 1000)).dot(P_inv)
print("A^1000 =")
print(A_1000)

# Tính lũy thừa D^2 (ma trận đường chéo)
print("D^2 =")
print(np.linalg.matrix_power(D, 2))
```

```python
#Câu 2: Sinh viên hãy viết các chương trình tính toán theo các công thức: 
import numpy as np
import math

# Nhập 2 vector u, v (bạn có thể thay bằng input nếu muốn)
u = np.array([1, 2, 3])
v = np.array([4, 5, 6])

# a. Tính khoảng cách giữa 2 vector u và v: ||u - v||
khoang_cach = np.linalg.norm(u - v)
print("Khoảng cách giữa u và v:", khoang_cach)

# b. Tính góc θ giữa 2 vector u và v:
# θ = arccos( <u, v> / (||u|| * ||v||) )
cos_theta = np.inner(u, v) / (np.linalg.norm(u) * np.linalg.norm(v))
# Đảm bảo cos_theta trong khoảng [-1,1] để tránh lỗi do sai số làm tròn
cos_theta = max(min(cos_theta, 1), -1)
theta_rad = math.acos(cos_theta)  # góc theo radian
theta_deg = math.degrees(theta_rad)  # góc theo độ
print("Góc giữa u và v (radian):", theta_rad)
print("Góc giữa u và v (độ):", theta_deg)

# c. Phép chiếu trực giao p của u trên v:
# p = (<u, v> / <v, v>) * v
p = (np.inner(u, v) / np.inner(v, v)) * v
print("Phép chiếu trực giao p của u trên v:", p)
```
