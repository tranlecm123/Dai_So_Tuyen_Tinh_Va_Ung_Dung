```python
# Bai 3
import numpy as np
from numpy import linalg as LA

# Định nghĩa ma trận A và B
A = np.matrix([[1, 0], [0, 0]])
B = np.matrix([[0, 0], [0, 1]])

# Phương pháp 1: Sử dụng lệnh inv
try:
    LA.inv(A)
except Exception as e:
    print("LA.inv(A):", e)  # Lỗi cho ma trận không khả nghịch

try:
    LA.inv(B)
except Exception as e:
    print("LA.inv(B):", e)  # Lỗi cho ma trận không khả nghịch

try:
    result = LA.inv(A + B)
    print("LA.inv(A+B):", result)  # Kết quả nếu khả nghịch
except Exception as e:
    print("LA.inv(A+B):", e)

# Phương pháp 2: Sử dụng lệnh det
print("LA.det(A):", LA.det(A))  # Định thức của A
print("LA.det(B):", LA.det(B))  # Định thức của B
print("LA.det(A+B):", LA.det(A + B))  # Định thức của A + B
```

```python
# Bai 4
import sympy as sp

# Khai báo 2 biến x và y
x, y = sp.symbols('x y')
A = sp.Matrix([[x, y], [y, x]])

# Khai báo các biến x1, y1, x2, y2
x1, y1 = sp.symbols('x1 y1')
A1 = sp.Matrix([[x1, y1], [y1, x1]])
x2, y2 = sp.symbols('x2 y2')
A2 = sp.Matrix([[x2, y2], [y2, x2]])

# Tính A1.T
print("A1.T:")
print(A1.T)

# Tính (A1+A2).T
print("(A1+A2).T:")
print((A1 + A2).T)

# Kiểm tra ((A1+A2).T).equals(A1+A2)
print("((A1+A2).T).equals(A1+A2):")
print(((A1 + A2).T).equals(A1 + A2))

# Khai báo hằng số c
c = sp.symbols('c')
print("c*A:")
print(c * A)

# Kiểm tra ((c*A).T).equals(c*A)
print("((c*A).T).equals(c*A):")
print(((c * A).T).equals(c * A))
```

```python
# Bai 7
import sympy as sp

# Khai báo các biến
c1, c2 = sp.symbols('c1 c2')
u1, u2 = sp.symbols('u1 u2')

print("=== KIỂM TRA TÍNH TẬP SINH ===")
# Giải hệ: c1 + c2 = u1, c1 - c2 = u2
he_phuong_trinh = [c1 + c2 - u1, c1 - c2 - u2]
nghiem_tap_sinh = sp.solve(he_phuong_trinh, (c1, c2))
print("Nghiệm tổng quát theo u1, u2:")
print(nghiem_tap_sinh)

print("\n=== KIỂM TRA TÍNH ĐỘC LẬP TUYẾN TÍNH ===")
# Thay u1 = 0, u2 = 0 để kiểm tra độc lập tuyến tính
he_dlt = [c1 + c2 - 0, c1 - c2 - 0]
nghiem_dlt = sp.solve(he_dlt, (c1, c2))
print("Nghiệm khi u1 = 0, u2 = 0:")
print(nghiem_dlt)

# Kết luận
if nghiem_dlt == {c1: 0, c2: 0}:
    print("\n=> Hai vector (1,1) và (1,-1) độc lập tuyến tính.")
else:
    print("\n=> Hai vector (1,1) và (1,-1) không độc lập tuyến tính.")

print("\n=> KẾT LUẬN: Tập {(1,1), (1,-1)} là một cơ sở của R^2.")
```

```python
# 2. Ánh xạ tuyến tính
# 2.2.Bài toán ứng dụng: Ma trận biến đổi (ảnh/font chữ/…)
import numpy as np

# Khởi tạo điểm P gốc
P = np.array([
    [0, 0, 3, 3, 1, 1, 2, 2, 1, 1],
    [0, 5, 5, 4, 4, 3, 3, 2, 2, 0]
])

# --- a. Tịnh tiến (vecdelta = [4, 2])
vecdelta_a = np.array([4, 2])
P_caua = (P.T + vecdelta_a).T
print("Câu a - Tịnh tiến:")
print(P_caua)

# --- b. Tịnh tiến và co giãn: F = 1.0, G = 2.0, vecdelta = [4, -2]
vecdelta_b = np.array([4, -2])
matran_b = np.array([
    [1.0, 0.0],
    [0.0, 2.0]
])
P_caub = (P.T @ matran_b + vecdelta_b).T
print("\nCâu b - Tịnh tiến + Co giãn:")
print(P_caub)

# --- c. Tịnh tiến + đối xứng trục y: vecdelta = [4, -2], ma trận đối xứng y
vecdelta_c = np.array([4, -2])
matran_c = np.array([
    [-1, 0],
    [0, 1]
])
P_cauc = (P.T @ matran_c + vecdelta_c).T
print("\nCâu c - Tịnh tiến + Đối xứng theo trục y:")
print(P_cauc)

# --- d. Tịnh tiến + shearing: L = 1, M = 0, vecdelta = [4, -2]
vecdelta_d = np.array([4, -2])
matran_d = np.array([
    [1, 1],  # L = 1
    [0, 1]   # M = 0
])
P_caud = (P.T @ matran_d + vecdelta_d).T
print("\nCâu d - Tịnh tiến + Shearing:")
print(P_caud)

# --- e. Đối xứng qua y = 0: x → -x, y giữ nguyên
matran_e = np.array([
    [-1, 0],
    [0, 1]
])
P_sym_y0 = (P.T @ matran_e).T
print("\nĐối xứng qua đường thẳng y = 0:")
print(P_sym_y0)

# --- f. Đối xứng qua đường thẳng y = x: hoán đổi x, y
matran_f = np.array([
    [0, 1],
    [1, 0]
])
P_sym_y_eq_x = (P.T @ matran_f).T
print("\nĐối xứng qua đường thẳng y = x:")
print(P_sym_y_eq_x)
```

```python
##### BAI TAP CHUONG 08 ########
#Bai1:
import sympy as sp

# Khai báo các biến hình thức
x, y, z = sp.symbols('x y z')  # Biến cho vector đầu vào
a, b, c = sp.symbols('a b c')  # Vector x = (a, b, c)
d, e, f = sp.symbols('d e f')  # Vector y = (d, e, f)
q = sp.symbols('q')  # Hằng số q cho kiểm tra tính đồng nhất

# Định nghĩa ánh xạ g(x, y, z) = (x + y, z + 2, 0)
g = lambda x1, x2, x3: (x1 + x2, x3 + 2, 0)

# Tách các thành phần của g
g1 = x + y  # Thành phần thứ nhất: x + y
g2 = z + 2  # Thành phần thứ hai: z + 2
g3 = 0      # Thành phần thứ ba: 0

# Tạo hàm lambdify để tính toán
f1 = sp.lambdify([x, y, z], g1, 'numpy')
f2 = sp.lambdify([x, y, z], g2, 'numpy')
f3 = sp.lambdify([x, y, z], g3, 'numpy')

# Kiểm tra điều kiện 1: g(qx) = q g(x)
# Tính g(q * (a, b, c))
g_qx = g(q * a, q * b, q * c)
# Tính q * g(a, b, c)
g_x = g(a, b, c)
q_g_x = (q * g_x[0], q * g_x[1], q * g_x[2])

# So sánh từng thành phần
cond1 = (g_qx[0].equals(q_g_x[0]) and
         g_qx[1].equals(q_g_x[1]) and
         g_qx[2].equals(q_g_x[2]))

print("Điều kiện 1 (g(qx) = q g(x)): ", cond1)

# Kiểm tra điều kiện 2: g(x + y) = g(x) + g(y)
# Tính g((a, b, c) + (d, e, f))
g_x_plus_y = g(a + d, b + e, c + f)
# Tính g(a, b, c) + g(d, e, f)
g_x = g(a, b, c)
g_y = g(d, e, f)
g_x_plus_g_y = (g_x[0] + g_y[0], g_x[1] + g_y[1], g_x[2] + g_y[2])

# So sánh từng thành phần
cond2 = (g_x_plus_y[0].equals(g_x_plus_g_y[0]) and
         g_x_plus_y[1].equals(g_x_plus_g_y[1]) and
         g_x_plus_y[2].equals(g_x_plus_g_y[2]))

print("Điều kiện 2 (g(x + y) = g(x) + g(y)): ", cond2)

# Kết luận
if cond1 and cond2:
    print("Ánh xạ g là ánh xạ tuyến tính.")
else:
    print("Ánh xạ g KHÔNG phải là ánh xạ tuyến tính.")
```
